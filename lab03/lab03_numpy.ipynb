Ćwiczenie 3: Operacje wektorowe — analiza biznesowa

Dane

```python
# Produkty w sklepie internetowym
nazwy = ['Laptop', 'Tablet', 'Smartfon', 'Słuchawki', 'Ładowarka',
'Klawiatura', 'Mysz', 'Monitor', 'Kamera', 'Drukarka']
ceny = np.array([3499, 1899, 2499, 349, 89, 249, 129, 1499, 599, 899])
ilosci_magazyn = np.array([15, 42, 78, 156, 312, 89, 201, 23, 45, 34])
oceny_klientow = np.array([4.7, 4.2, 4.5, 4.8, 3.9, 4.1, 4.3, 4.6, 3.8, 3.5])
```

Zadanie 1: Podstawowe statystyki cenowe

```python
# Średnia cena, najdroższy produkt, najtańszy produkt
# Podpowiedzi: .mean(), .max(), .min(), .argmax(), .argmin()
srednia_cena = ceny.mean()
najdrozszy_produkt = ceny.argmax()
najtanszy_produkt = ceny.argmin()
print(srednia_cena)
print(nazwy[najdrozszy_produkt], ceny[najdrozszy_produkt])
print(nazwy[najtanszy_produkt], ceny[najtanszy_produkt])
```

Średnia cena produktów wynosi 1171zł. Najdroższy produkt to laptop (3499zł) natomiast najtańszy to ładowarka (89zł).

Zadanie 2: Wartość magazynu

```python
# Oblicz wartość każdego produktu (cena × ilość) i łączną wartość magazynu
# Podpowiedź: mnożenie tablic element-wise
wartosci = ceny * ilosci_magazyn
print(wartosci)
print(wartosci.sum())
```

Wartości produktów: 52485zł, 79758zł, 194922zł, 54444zł, 27768zł, 22161zł,  25929zł, 34477zł, 26955zł, 30566zł. 

Łączna wartóść magazynu: 549465zł.

Zadanie 3: Analiza cenowa

```python
# Stworz array z cenami po rabacie 15% (promocja świąteczna)
# Stworz array z cenami po uwzględnieniu VAT 23%
# Ile produktów kosztuje powyżej 1000 zł?
# Podpowiedź: rabat = ceny * 0.85, VAT = ceny * 1.23, zliczanie: (warunek).sum()
rabat = ceny * 0.85
vat = ceny * 1.23
powyzej1000 = ceny > 1000
print(rabat)
print(vat)
print(powyzej1000.sum())
```

Ceny produktów objętnych promocją świąteczną (-15%): 2974.15zł, 1614.15zł, 2124.15zł, 296.65zł, 75.65zł, 211.65zł, 109.65zł, 1274.15zł, 509.15zł, 764.15zł.

Ceny produktów po uwzględnieniu VAT(23%): 4303.77zł, 2335.77zł, 3073.77zł, 429.27zł, 109.47zł, 306.27zł, 158.67zł, 1843.77zł, 736.77zł, 1105.77zł.

Ilość produktów, których cena wynosi więcej niż 1000zł jest równa 4.


Zadanie 4: Filtrowanie

```python
# Produkty z oceną >= 4.5 — nazwy i ceny
# Produkty poniżej 200 zł — nazwy i ilości w magazynie
# Podpowiedź: maska = oceny_klientow >= 4.5, potem ceny[maska]
# Uwaga: nazwy to zwykła lista — zamień ją na np.array(nazwy), żeby użyć maski
maska = oceny_klientow >= 4.5
nazwy = np.array(nazwy)
print("Produkty z oceną >= 4.5: ", nazwy[maska], ceny[maska])

ponizej200 = ceny < 200
print("Produkty poniżej 200zł :", nazwy[ponizej200], ilosci_magazyn[ponizej200])
```

Produkty z oceną >= 4.5: Laptop (3499zł), Smartfon(2499zł), Słuchawki(349zł),Monitor(1499zł).

Produnkty, których cena wynosi mniej niż 200zł: Ładowarka(312szt.), Mysz(201szt.)

Zadanie 5: Analiza zapasów

```python
# Produkty z niskim stanem magazynowym (< 30 szt.) — nazwy
# Łączna wartość tych produktów
# Ile % całego magazynu stanowią?
# Podpowiedź: wartosci = ceny * ilosci_magazyn, potem filtruj maską
wartosci = ceny * ilosci_magazyn
maska = ilosci_magazyn < 30
wartosc_niski = wartosci[maska].sum()
procent = wartosc_niski / wartosci.sum() * 100

print(nazwy[maska])
print(wartosc_niski)
print(procent)
```

Produkty z niskim stanem magazynowym (<30szt.): Laptop, Monitor.

Łączna watość tych produktów wynosi: 86962zł.

Podane produkty stanowią ok.15.83% całego magazynu

Zadanie 6 (bonus): Analiza korelacji cena–ocena

```python
# Które produkty są "okazją" — ocena >= 4.5 i cena < 1000 zł? Wypisz nazwy i ceny.
# Jaka jest średnia cena produktów z oceną >= 4.0 vs < 4.0?
# Podpowiedź: łączenie warunków: (warunek1) & (warunek2)
okazja = (oceny_klientow >= 4.5) & (ceny < 1000)
print(nazwy[okazja], ceny[okazja])

oceny_dobre = oceny_klientow >= 4.0
oceny_slabe = oceny_klientow < 4.0
srednia_dobre = ceny[oceny_dobre].mean()
srednia_slabe = ceny[oceny_slabe].mean()
print(srednia_dobre)
print(srednia_slabe)
```

Okazja: Słuchawki 

Cena: 349zł

Średnia cena produktów z oceną >= 4.0 wynosi ok.1446.14zł.

Średnia cena produktów z oceną < 4.0 wynosi 529zł.

Ćwiczenie 4: Benchmark — lista vs NumPy

```python
#Benchmark
import time
```

```python
# Przygotowanie danych
rozmiar = 1_000_000
lista = list(range(rozmiar))
tablica = np.array(lista)
```

```python
# Test 1: Pomnóż przez 2
start = time.perf_counter()
wynik_lista = [x * 2 for x in lista]
czas_lista = time.perf_counter()- start
start = time.perf_counter()
wynik_numpy = tablica * 2
czas_numpy = time.perf_counter()- start
print(f"=== Mnożenie ×2 ({rozmiar:,} elementów) ===")
print(f"Lista: {czas_lista*1000:.1f} ms")
print(f"NumPy: {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

```python
# Test 2: Suma
start = time.perf_counter()
suma_lista = sum(lista)
czas_lista = time.perf_counter()- start
start = time.perf_counter()
suma_numpy = tablica.sum()
czas_numpy = time.perf_counter()- start
print(f"\n=== Suma ({rozmiar:,} elementów) ===")
print(f"Lista: {czas_lista*1000:.1f} ms")
print(f"NumPy: {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

```python
# Test 3: Filtrowanie (elementy > 500000)
start = time.perf_counter()
filtr_lista = [x for x in lista if x > 500000]
czas_lista = time.perf_counter()- start
start = time.perf_counter()
filtr_numpy = tablica[tablica > 500000]
czas_numpy = time.perf_counter()- start
print(f"\n=== Filtrowanie > 500k ({rozmiar:,} elementów) ===")
print(f"Lista: {czas_lista*1000:.1f} ms")
print(f"NumPy: {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

## Wnioski z benchmarku
- NumPy jest **33×** szybszy w mnożeniu
- NumPy jest **15×** szybszy w sumowaniu
- NumPy jest **27×** szybszy w filtrowaniu
- Wniosek: przy dużych danych **zawsze** używaj NumPy zamiast list
